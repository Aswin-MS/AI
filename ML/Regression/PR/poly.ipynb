```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('poly.csv')
```

```python
df
```

```python
df.drop(['sno'],axis=1,inplace=True)
```

```python
df.dtypes
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x,y)
y_pred=model.predict(x)
y_pred
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred)
```

```python
from sklearn.metrics import mean_absolute_error,mean_absolute_percentage_error,mean_squared_error,root_mean_squared_error,r2_score
print('MAE: ',mean_absolute_error(y,y_pred))
print('MAPE: ',mean_absolute_percentage_error(y,y_pred))
print('MSE: ',mean_squared_error(y,y_pred))
print('RMSE: ',root_mean_squared_error(y,y_pred))
print('R2: ',r2_score(y,y_pred))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
x_poly
```

```python
x_poly.shape
```

```python
model1=LinearRegression()
model1.fit(x_poly,y)
y_pred1=model1.predict(x_poly)
y_pred1
```

```python
plt.scatter(x,y)
plt.plot(x,y_pred1)
```

```python
print('MAE: ',mean_absolute_error(y,y_pred1))
print('MAPE: ',mean_absolute_percentage_error(y,y_pred1))
print('MSE: ',mean_squared_error(y,y_pred1))
print('RMSE: ',root_mean_squared_error(y,y_pred1))
print('R2: ',r2_score(y,y_pred1))
```
