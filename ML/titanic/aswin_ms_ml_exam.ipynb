1.	What is the survival rate for males vs. females?


```python
q1=df[['Sex','Survived']].groupby('Sex',as_index=False).mean().sort_values(by='Survived',ascending=False)
q1
```

1) Ans: Femal survival rate is 0.74 compared to male survival rate of 0.18, this shows about 75% of survived are women.

2.	Did passengers in higher classes (1st class) have a higher survival rate compared to lower classes (3rd class)?

```python
q2=df[['Pclass','Survived']].groupby('Pclass',as_index=False).mean().sort_values(by='Survived',ascending=False)
q2
```

2) Ans: Yes, passengers in higher classes(63%) survived more than lower classes(24%).

3.	Did passengers with more family members (siblings/spouses or parents/children) aboard have a higher or lower chance of survival?

```python
q3=df[['SibSp','Survived']].groupby('SibSp',as_index=False).mean().sort_values(by='Survived',ascending=False)
q3
```

```python
q31=df[['Parch','Survived']].groupby('Parch',as_index=False).mean().sort_values(by='Survived',ascending=False)
q31
```

3) Ans: From the above observation, we can see that passengers with more number of siblings/spouses were less likely to survive with 'SIbSp'= 5,8 having 0 chance of survival.
The same applies for passengers having parents/childrens, with greather number like 4 and 6 having 0 survived.
So, passengers with more family memebers were less likely to survive.

4.	Did the port of embarkation (C, Q, S) influence the survival rate?

```python
q4=df[['Embarked','Survived']].groupby('Embarked',as_index=False).mean().sort_values(by='Survived',ascending=False)
q4
```

5.	Did passengers who paid higher fares have a better chance of survival?

```python
q5=df[['Fare','Survived']].groupby('Fare',as_index=False).mean().sort_values(by='Fare',ascending=False)
q5
```

```python
from sklearn.preprocessing import LabelEncoder
lab=LabelEncoder()
df['Sex']=lab.fit_transform(df['Sex'])
df['Ticket']=lab.fit_transform(df['Ticket'])
df['Embarked']=lab.fit_transform(df['Embarked'])
```

```python
df
```

```python
x=df.drop('Survived',axis=1)
x.ndim
```

```python
y=df['Survived']
y.ndim
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.30,random_state=0)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
score=accuracy_score(y_test,y_pred)
score
```

```python
print(classification_report(y_test,y_pred))
```
