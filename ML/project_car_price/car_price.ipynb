Name column has 1878 unique values , so we can't visualize it.

We don't need to encode columns like 'Mileage','Engine', and 'Power' because it has numerical value, we need to remove the 'kmpl' or 'CC' or 'bhp' from it.

We will be using 'get dummies' encoding for all columns except 'Name' column.
'Name' column will be splitted to take the company name.

```python
#Splitting the name based on Brands
df_train['Brand']=df_train['Name'].apply(lambda x:x.split(' ')[0])
df_train
```

```python
df_train['Brand'].unique()
```

```python
c=df_train['Brand'].value_counts()
c
```

```python
plt.figure(figsize=(8,8))
plt.bar(c.index,c.values,color='gray')
plt.title('Brand Count')
plt.xlabel('Brand')
plt.xticks(rotation=90)
plt.ylabel('Count')
```

Now we encode 'Brand' also.

```python
#get_dummies Encoding
dfe=pd.get_dummies(df_train[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True,dtype=int)
dfe

#dtype= int --> gives output as 1/0
#drop_first=True ---> to drop the first column as it is not necessary
```

We combine the df_train and dfe by concat()

```python
df1=pd.concat([df_train,dfe],axis=1)
df1
#axis=1 for column
```
