Now we encode 'Brand' also.

```python
#get_dummies Encoding
dfe=pd.get_dummies(df_train[['Location','Fuel_Type','Transmission','Owner_Type','Brand']],drop_first=True,dtype=int)
dfe

#dtype= int --> gives output as 1/0
#drop_first=True ---> to drop the first column as it is not necessary inorder to reduce the number of columns
```

We combine the df_train and dfe by concat()

```python
df1=pd.concat([df_train,dfe],axis=1)
df1
#axis=1 for column
```

48+14 columns

```python
df1.shape
```

```python
df_train.columns
```

```python
#now we drop columns
#we drop New_Price also because it has less values, most are missing value
df1.drop(['Unnamed: 0','Name','Location','Fuel_Type','Transmission','Owner_Type','New_Price','Brand'],axis=1,inplace=True)
```

Now we have 55 columns in total.

```python
#String Replacements of Power,Mileage,Engine
df1['Mileage']=df1['Mileage'].str.replace('km/kg','')
df1['Mileage']=df1['Mileage'].str.replace('kmpl','')
df1['Engine']=df1['Engine'].str.replace('CC','')
df1['Power']=df1['Power'].str.replace('bhp','')
```

```python
df1
```

In some column some values are entered as 'null', we need to replace it with '0' as the datatype is object, we need to pass 0 as string.

```python
df1['Power'].unique()
```

```python
df1['Mileage'].unique()
```

There is a value passed as 'null', we need to replace it with '0'.

```python
#null replacement
df1['Power']=df1['Power'].str.replace('null','0')
df1['Engine']=df1['Engine'].str.replace('null','0')
df1['Mileage']=df1['Mileage'].str.replace('null','0')
```

```python
df1['Power'].unique()
```

We need to convert the datatype of 'Mileage','Power','Engine' as they are now objects.

```python
#Data Type Convertion
df1['Mileage']=df1['Mileage'].astype(float)
df1['Power']=df1['Power'].astype(float)
df1['Engine']=df1['Engine'].astype(float)
```

```python
df1.dtypes
```

No vehicle will have power, mileage or engine with '0', so we treat that '0' as a missing value, as we changed the 'null' value to '0' and later converted it to float datatype, we still need to change 0 to a value that can be suitable for those columns.

Mileage---->num----->0----->treat as missing value

Now we perform Numerical Data Replacement

We change all of them as Nan.

Assign np.nan to elements within a NumPy array to indicate missing values.

```python
#numerical data replacement
#We replace 0 with nan
df1.loc[df1.Engine==0,'Engine']=np.nan
df1.loc[df1.Power==0,'Power']=np.nan
df1.loc[df1.Mileage==0,'Mileage']=np.nan
```

```python
df1.isna().sum()
```

```python
df1['Mileage']=df1['Mileage'].fillna(df1['Mileage'].median())
df1['Engine']=df1['Engine'].fillna(df1['Engine'].mean())
df1['Power']=df1['Power'].fillna(df1['Power'].mean())
df1['Seats']=df1['Seats'].fillna(df1['Seats'].mode() [0])
```
