```python
import requests
import csv
from bs4 import BeautifulSoup
url="https://www.amazon.in/s?k=phones&crid=QE5E10XQXLNZ&sprefix=phones%2Caps%2C235&ref=nb_sb_noss_2"
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)",
    "Accept-Language": "en-US,en;q=0.5"
}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")
soup=BeautifulSoup(response.text,'html.parser')
h=soup.prettify()
```

```python
print(h)
```

```python
products = soup.find_all("div", {"data-component-type": "s-search-result"})
print(products)
```

```python
product_list = []

for product in products:
    # Title
    title = product.find("span", class_="a-size-medium")
    title = title.text.strip() if title else "N/A"
    
    # Price
    price = product.find("span", class_="a-price-whole")
    price = price.text.strip() if price else "N/A"
    
    # Rating
    rating = product.find("span", class_="a-icon-alt")
    rating = rating.text.strip() if rating else "N/A"
    
    # Reviews
    reviews = product.find("span", class_="a-size-base")
    reviews = reviews.text.strip() if reviews else "N/A"
    
    # Product Link
    link_tag = product.find("a", class_="a-link-normal")
    link = "https://www.amazon.in" + link_tag["href"] if link_tag else "N/A"
    
    # Save the product
    product_list.append([title, price, rating, reviews, link])

# Save to CSV
with open("amazon_scraped_page.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["Title", "Price", "Rating", "Reviews", "Product Link"])
    writer.writerows(product_list)
print("Scraping completed! File saved as amazon_scraped_page.csv")
```
