```python
import cv2
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r'D:\apps\Tesseract ocr\tesseract.exe'
```

```python
img=cv2.imread('car1.jpg')
img
```

```python
cv2.imshow('Car',img)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
data=pytesseract.image_to_string(img)
print(data)
```

```python
prop=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
prop
```

```python
n=len(prop['text'])
n
```

```python
for i in range(n):
    if prop['text'][i].strip()!='':
        x=prop['left'][i]
        y=prop['top'][i]
        w=prop['width'][i]
        h=prop['height'][i]
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),2)
cv2.imshow('car_number_det',img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
