```python
import cv2
import pytesseract
pytesseract.pytesseract.tesseract_cmd=r'D:\apps\Tesseract ocr\tesseract.exe'
```

```python
jd=[' Python',' R', 'SQL', 'Pyspark','Pandas', 'NumPy', 'Scikit-learn', 'TensorFlow', 'PyTorch', 'Keras','SQL', 'MySQL', 'Hadoop', 'Spark', 'Hive', 'MongoDB', 'AWS', 'Azure', 'GCP','Tableau', 'Power BI','Matplotlib', 'Seaborn', 'ggplot' ]
jd
```

```python
l=len(jd)
l
```

```python
img=cv2.imread('resume.png')
cv2.imshow('Resume',img)
cv2.waitKey(3000)
cv2.destroyAllWindows()
```

```python
data=pytesseract.image_to_string(img)
print(data)
```

```python
prop=pytesseract.image_to_data(img,output_type=pytesseract.Output.DICT)
prop
```

```python
n=len(prop['text'])
n
```

```python
skill=[skill.lower().strip() for skill in jd]
jd_skill=[word.lower().strip() for word in prop['text'] if prop['text']!= '' ]
```

```python
match=set()
```

```python
score=0
for i in range(n):
    for j in range(l):
        if jd_skill[i]==skill[j]:
            match.add(prop['text'][i])
score=len(match)
print(score)
print(match)
```

```python
score_per=(score/l)*100
score_per
```

```python
score_list=list(match)
score_list
```

```python
for i in range(n):
    for j in range(len(score_list)):
        if prop['text'][i]==score_list[j]:
            x=prop['left'][i]
            y=prop['top'][i]
            w=prop['width'][i]
            h=prop['height'][i]
            cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,255),1)
cv2.imshow("image2",img)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
