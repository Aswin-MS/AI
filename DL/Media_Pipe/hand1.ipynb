```python
import mediapipe as mp
import cv2
```

## Hand Landmark

```python
mp_hand=mp.solutions.hands
mp_draw=mp.solutions.drawing_utils
```

```python
data=mp_hand.Hands() #class
```

```python
video=cv2.VideoCapture(0)
while True:
    succ,img=video.read()
    frame=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(frame)
    print(result.multi_hand_landmarks)
    if result.multi_hand_landmarks: #if hand is detected
        for x in result.multi_hand_landmarks:
            mp_draw.draw_landmarks(img,x,mp_hand.HAND_CONNECTIONS,mp_draw.DrawingSpec(color=(0,0,255),thickness=5,circle_radius=5),mp_draw.DrawingSpec(color=(255,0,0),thickness=5))


    cv2.imshow('Webcam',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
