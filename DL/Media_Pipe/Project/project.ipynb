```python
import mediapipe as mp
import cv2
```

```python
mp_hand=mp.solutions.hands
mp_draw=mp.solutions.drawing_utils
```

```python
data=mp_hand.Hands(max_num_hands=1)
```

```python
video=cv2.VideoCapture(0)
while True:
    succ,img=video.read()
    frame=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(frame)
    #print(result.multi_hand_landmarks)
    tipid=[4,8,12,16,20]
    lmlst=[]
    if result.multi_hand_landmarks:
        for x in result.multi_hand_landmarks:
            for id,lm in enumerate(x.landmark):
                cx=lm.x
                cy=lm.y
                lmlst.append([id,cx,cy])
                # print(lmlst)
                if len(lmlst)!=0 and len(lmlst)==21: #only work after detecting all 5 fingers
                    fingercount=[]
                    #thumb
                    if lmlst[12][1]< lmlst[20][1]: #to check which hand
                        if lmlst[4][1] >lmlst[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    else:
                        if lmlst[4][1]<lmlst[3][1]:
                            fingercount.append(1)
                        else:
                            fingercount.append(1)

                    for i in range(1,5): #i=1,2,3,4 --> all finger except thumb
                        if lmlst[tipid[i]][2]>lmlst[tipid[i]-2][2]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    #print(fingercount)
                    if len(fingercount)!=0:
                        fingercount1=fingercount.count(1)
                    #print(fingercount1)
                    count=str(fingercount1)
                    cv2.putText(img,"Count"+count,(35,400),cv2.FONT_HERSHEY_COMPLEX,3,(255,0,0))
            mp_draw.draw_landmarks(img,x,mp_hand.HAND_CONNECTIONS)
        cv2.imshow('Finger_Count',img)
        if cv2.waitKey(1) & 0XFF==ord('q'):
            break
video.release()
cv2.destroyAllWindows()
```
