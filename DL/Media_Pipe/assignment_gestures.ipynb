```python
import mediapipe as mp
import cv2
```

```python
mp_hand=mp.solutions.hands
mp_draw=mp.solutions.drawing_utils
```

```python
data=mp_hand.Hands(max_num_hands=1)  
```

```python
video=cv2.VideoCapture(0)
while True:
    suc,img=video.read()
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    result=data.process(img1)
    tip_id=[4,8,12,16,20]
    lmlist=[]



    if result.multi_hand_landmarks:  
        for handlms in result.multi_hand_landmarks:
            for id,lm in enumerate(handlms.landmark):
                cx=lm.x
                cy=lm.y
                lmlist.append([id,cx,cy])
                if len(lmlist)!=0 and len(lmlist)==21:
                    fingercount=[]
                    #thumbs code:
                    if lmlist[12][1] < lmlist[20][1]:
                        if lmlist[4][1] > lmlist[3][1]: 
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                    else:
                        if lmlist[4][1] < lmlist[3][1]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)



                    for i in range(1,5):  #except thumb
                        if lmlist[tip_id[i]][2] > lmlist[tip_id[i]-2][2]:
                            fingercount.append(0)
                        else:
                            fingercount.append(1)
                   
                    gesture="Unknown"
                    if fingercount==[0,0,0,0,0]:
                        gesture="Fist"
                    elif fingercount==[1,1,1,1,1]:
                        gesture="Open Palm"
                    elif fingercount==[0,1,0,0,0]:
                        gesture="Pointing"
                    elif fingercount==[0,1,1,0,0]:
                        gesture="Peace sign" 
                    elif fingercount==[1,0,0,0,0]:
                        gesture="Thumbs up"
                    elif fingercount==[0,0,0,0,1]:
                        gesture="one"
                    elif fingercount==[1,1,0,0,0]:
                        gesture="L"
                    elif fingercount==[1,0,0,0,1]:
                        gesture="Call Me"
                    elif fingercount==[0,1,1,1,0]:
                        gesture="Three"
                    elif fingercount==[1,1,1,0,0]:
                        gesture="Gun Point"
                    elif fingercount==[0,0,1,1,1]:
                        gesture="okay!"

            mp_draw.draw_landmarks(img,handlms,mp_hand.HAND_CONNECTIONS,mp_draw.DrawingSpec(color=(0,0,255),thickness=5,circle_radius=5),mp_draw.DrawingSpec(color=(255,0,0),thickness=5))   #to here
            cv2.putText(img,"Gesture: "+str(gesture),(0,400),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)


    cv2.imshow('Finger Count',img)
    if cv2.waitKey(1)&0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
