```python
img2=cv2.imread('children_image\c2.jpg')
img3=cv2.imread('children_image\c3.jpg')
img4=cv2.imread('children_image\c4.jpg')
```

```python
cv2.imshow('image_display1',img2)
cv2.imshow('image_display2',img3)
cv2.imshow('image_display3',img4)
cv2.waitKey(7000)
cv2.destroyAllWindows()
```

```python
#convert image into grayscale
gray_img=cv2.imread('children_image\c1.jpg',cv2.IMREAD_GRAYSCALE)
gray_img
```

```python
cv2.imshow('image_display_gray',gray_img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

```python
#save an image
cv2.imwrite('children_image/c1gray.jpg',gray_img)
```

```python
cv2.imwrite('children_image/c1gray.png',gray_img)
```

```python
#convert to binary image
#color-->grayscale-->binary
thresh,bin_img=cv2.threshold(gray_img,128,255,cv2.THRESH_BINARY)
bin_img
```

```python
#binary image display
cv2.imshow('Binaryimage',bin_img)
cv2.waitKey(7000)
cv2.destroyAllWindows()
```

```python
#grayimage 2
gray_img1=cv2.imread('children_image/c2.jpg',cv2.IMREAD_GRAYSCALE)
thresh,bin_img2=cv2.threshold(gray_img1,160,255,cv2.THRESH_BINARY)
cv2.imshow('binary_image2',bin_img2)
cv2.waitKey(5000)
cv2.destroyAllWindows()
```
