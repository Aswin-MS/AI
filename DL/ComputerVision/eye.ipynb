```python
import cv2
import matplotlib.pyplot as plt
```

```python
eye=cv2.CascadeClassifier('haarcascade_eye.xml')
```

```python
img=cv2.imread('children_image/c1.jpg')
img
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
eyes=eye.detectMultiScale(conv_img,minNeighbors=7)
eyes
```

```python
for x,y,w,h in eyes:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),1)
plt.imshow(conv_img)
```

```python
#video
video=cv2.VideoCapture(0)
#use loop to capture video
while True:
    success,img=video.read()  #returns 2 variable
    img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    eyes=eye.detectMultiScale(img1,minNeighbors=6)
    for x,y,w,h in eyes:
        cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    cv2.imshow('webcam',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
