```python
import cv2
import ultralytics
```

```python
model=ultralytics.YOLO('yolov8n.pt')
```

```python
video=cv2.VideoCapture(0)
while True:
    succ,img=video.read()
    result=model(img)
    for i in result:
        i .save('result1.jpg')
        img=cv2.imread('result1.jpg')
        cv2.imshow('image1',img)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
