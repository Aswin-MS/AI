```python
import cv2
import ultralytics
```

```python
model=ultralytics.YOLO('yolov8n.pt')
```

```python
result=model(['image.jpg','images.jpg'])
```

```python
for i in result:
    i.save('result.jpg')
    img=cv2.imread('result.jpg')
    cv2.imshow("image",img)
    cv2.waitKey(10000)
    cv2.destroyAllWindows()
```
