```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('img/10face.jpg')
img
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_det=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
face=face_det.detectMultiScale(conv_img,minNeighbors=4)
face
```

```python
for x,y,w,h in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(255,0,0),2)
plt.imshow(conv_img)
```
