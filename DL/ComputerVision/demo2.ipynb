```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=cv2.imread('children_image/c4.jpg')
img
```

```python
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

Bounding Box 

*Rectangle, Circle

```python
#Rectangle
img_box1=cv2.rectangle(conv_img,(70,30),(145,115),color=(255,0,0),thickness=1)
plt.imshow(img_box1)
```

```python
#Circle
conv_img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img1)
```

```python
cv2.circle(conv_img1,(110,70),(45),color=(0,0,255),thickness=1)
plt.imshow(conv_img1)
```

```python
img2=cv2.imread('children_image/c3.jpg')
plt.imshow(img2)
```

```python
conv_img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img2)
```

```python
cv2.rectangle(conv_img2,(90,30),(150,100),color=(0,255,0),thickness=1)
plt.imshow(conv_img2)
```

```python
cv2.circle(conv_img2,(120,65),(35),color=(255,0,0),thickness=-1)
plt.imshow(conv_img2)
```

```python
img3=cv2.imread('children_image/c1.jpg')
conv_img3=cv2.cvtColor(img3,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img3)
```

```python
cv2.putText(conv_img3,text='happy',org=(100,50),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=2,color=(255,0,0),thickness=1)
plt.imshow(conv_img3)
```

```python
img4=cv2.imread('children_image/c2.jpg')
conv_img4=cv2.cvtColor(img4,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img4)
```

```python
cv2.rectangle(conv_img4,(40,40),(115,130),color=(255,0,0),thickness=1)
cv2.putText(conv_img4,'Boy',(0,20),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=0.8,color=(20,100,255),thickness=2)
plt.imshow(conv_img4)
```
